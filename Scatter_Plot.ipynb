```python
import pandas as pd 
import seaborn as sns
import numpy as np 
from matplotlib import pyplot as plt


df = pd.read_csv('Rbframe.csv')
df.drop(['Age'], axis=1, inplace=True)
df[['Fantasy Points Per Game (2019)']] = df[['Fantasy Points Per Game (2019)']].apply(pd.to_numeric)
df[['Weighted Opportunities Per Game (2019)']] = df[['Fantasy Points Per Game (2019)']].apply(pd.to_numeric)
df['Weighted Opportunities Per Game (2019)'] = df['Weighted Opportunities Per Game (2019)'].round(2)
df = df[df['Fantasy Points Per Game (2019)'] > 15 ]
df = df.sort_values(by=['Fantasy Points Per Game (2019)','College Dominator Rating'])

df.plot.scatter

N = 15
x = df['Fantasy Points Per Game (2019)']
y = df['Speed Score']
colors = np.random.rand(N)
area = (50 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)

plt.ylabel('Speed Score') 
plt.title('Fantasy Points Per Game vs Speed Score')

plt.xlabel('Fantasy Points Per Game (2019) ')

plt.show() 
df.head(25)

```
